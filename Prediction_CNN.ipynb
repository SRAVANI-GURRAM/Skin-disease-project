```python
from tensorflow.keras.models import load_model
import numpy as np
import cv2
import os

# Load the trained model
model = load_model("skin_cnn.h5")

# Define class labels with full names
class_labels = {
    'AK': 'Actinic Keratosis',
    'BCC': 'Basal Cell Carcinoma',
    'BKL': 'Benign Keratosis-like Lesions',
    'DF': 'Dermatofibroma',
    'MEL': 'Melanoma',
    'NV': 'Melanocytic Nevi',
    'SCC': 'Squamous Cell Carcinoma',
    'VASC': 'Vascular Lesions'
}

def preprocess_image(image_path):
    img = cv2.imread(image_path)
    if img is None:
        print(f"Error: Image not found at {image_path}")
        return None

    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # Convert to RGB
    img = cv2.resize(img, (28, 28))  # Resize to match training images
    img = img.astype('float32')  # Convert datatype
    img = img.reshape(-1, 28, 28, 3)  # Reshape for model input

    return img
```

```python
# Image path (Change to your image path)
image_path = "ISIC_0033494.jpg"

# Preprocess image
img = preprocess_image(image_path)

if img is not None:
    prediction = model.predict(img)
    predicted_index = np.argmax(prediction)  # Get index of highest probability
    predicted_short_label = list(class_labels.keys())[predicted_index]  # Get predicted short label
    predicted_label = class_labels[predicted_short_label]  # Get full name
    confidence_score = np.max(prediction) * 100  # Get confidence score in percentage

    print(f"Predicted Index: {predicted_index}")
    print(f"Predicted Class: {predicted_label}")
    print(f"Confidence Score: {confidence_score:.2f}%")
```
